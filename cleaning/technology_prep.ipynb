```python
# Creating 
import pandas as pd 
import numpy as np
import os
```

```python
#datadirectory = 

czone_data = pd.read_stata('H:/InnovationProject/data/temp/cz_preaggregate.dta')
czone_data = czone_data[czone_data['year']>=1988]
unique_year_values = czone_data['year'].unique()
#len(unique_year_values)
```

```python
czone_data.head()
```

```python
# Creating the loop for the larger czones 
data_directory = 'H:/InnovationProject/data/temp/czone/special_czones'
large_cz_values = [19600,20500,33100,37800]

for year in unique_year_values:
    for cz in large_cz_values:
        
        firstfilename = os.path.join(data_directory,f'maltec{cz}.dta')
        
        technology = pd.read_stata(firstfilename)
        czone_helper = czone_data [czone_data['czone'] == cz]
        czone_helper = czone_helper[czone_helper['year'] == year]
        
        merged_df = pd.merge(technology, czone_helper, on =['assignee_id_pat'])
        merged_df['sum_maltec'] = merged_df.groupby(['assignee_id'])['maltec'].transform('sum')

        merged_df['inventor_weight'] = merged_df['n_inventors3']/merged_df['sum_inventors']
        merged_df['tec_weight'] = merged_df['maltec']
        merged_df['technology_change1'] = merged_df['tec_weight'] * merged_df['change_other_threelargest']
        merged_df['technology_change2'] = merged_df['tec_weight'] * merged_df['inventor_weight'] * merged_df['change_other_threelargest']
        merged_df['weightedtec_change1'] = merged_df.groupby(['assignee_id'])['technology_change1'].transform('sum')
        merged_df['weightedtec_change2'] = merged_df.groupby(['assignee_id'])['technology_change2'].transform('sum')
        merged_df = merged_df.groupby(['assignee_id', 'czone', 'year']).first().reset_index()
        merged_df = merged_df.loc[:, ['assignee_id', 'czone', 'year', 'weightedtec_change1', 'weightedtec_change2']]
        
        final_directory = 'H:/InnovationProject/data/temp/czone/special_czones/tech_index'
        finalfilename = os.path.join(final_directory,f'tech_index{cz}_{year}year.dta')
        print(finalfilename)
        merged_df.to_stata(finalfilename)
        merged_df.head()
```

```python
# Creating the loop for the smaller czones  

folder_path = 'H:/InnovationProject/data/temp/czone'
for year in unique_year_values:
    for filename in os.listdir(folder_path):
        if filename.endswith('.dta'): 
            print(filename)
            file_path = os.path.join(folder_path, filename)
            
            cz = ''.join([char for char in filename if char.isdigit()])
            print(cz)
            cz = int(cz)
            
            technology = pd.read_stata(file_path)
            technology = technology.loc[:, ['assignee_id', 'assignee_id_pat', 'maltec']]
            czone_helper = czone_data [czone_data['czone'] == cz]
            czone_helper = czone_helper[czone_helper['year'] == year]
            merged_df = pd.merge(technology, czone_helper, on =['assignee_id_pat'])
            merged_df['sum_maltec'] = merged_df.groupby(['assignee_id'])['maltec'].transform('sum')
            merged_df['sum_inventors'] = merged_df.groupby(['assignee_id'])['n_inventors3'].transform('sum')
            merged_df['inventor_weight'] = merged_df['n_inventors3']/merged_df['sum_inventors']
            merged_df['tec_weight'] = merged_df['maltec']
            merged_df['technology_change1'] = merged_df['tec_weight'] * merged_df['change_other_threelargest']
            merged_df['technology_change2'] = merged_df['tec_weight'] * merged_df['inventor_weight'] * merged_df['change_other_threelargest']
            merged_df['weightedtec_change1'] = merged_df.groupby(['assignee_id'])['technology_change1'].transform('sum')
            merged_df['weightedtec_change2'] = merged_df.groupby(['assignee_id'])['technology_change2'].transform('sum')
            merged_df = merged_df.groupby(['assignee_id', 'czone', 'year']).first().reset_index()
            merged_df = merged_df.loc[:, ['assignee_id', 'czone', 'year', 'weightedtec_change1', 'weightedtec_change2']]
                
            final_directory = 'H:/InnovationProject/data/temp/czone/tech_index'
            finalfilename = os.path.join(final_directory,f'tech_index{cz}_{year}year.dta')
            print(finalfilename)
                
            merged_df.to_stata(finalfilename)    
```
